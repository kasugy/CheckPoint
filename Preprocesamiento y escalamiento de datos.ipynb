```python
# ============================================================
# ANALISIS DE CASO: PREPROCESAMIENTO Y ESCALAMIENTO DE DATOS
# RetailData Analytics
# ============================================================

# 1. IMPORTAR LIBRERIAS NECESARIAS
# ============================================================

import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, MinMaxScaler, StandardScaler

# ============================================================
# 2. CREAR EL DATAFRAME CON LOS DATOS PROPORCIONADOS
# ============================================================

datos = pd.DataFrame({
    'ClienteID': [1, 2, 3, 4],
    'Edad': [25, 45, 30, 40],
    'Ciudad': ['Madrid', 'Barcelona', 'Sevilla', 'Barcelona'],
    'Ingresos': [30000, 50000, np.nan, 40000]
})

print("=" * 60)
print("DATOS ORIGINALES")
print("=" * 60)
print(datos)
print("\n")

# ============================================================
# 3. IMPUTAR EL VALOR FALTANTE EN INGRESOS CON LA MEDIA
# ============================================================

print("=" * 60)
print("IMPUTACION DE VALORES FALTANTES")
print("=" * 60)

media_ingresos = datos['Ingresos'].mean()
print(f"Media de Ingresos: {media_ingresos}")

datos['Ingresos'] = datos['Ingresos'].fillna(media_ingresos)

print("\nDATOS CON VALOR FALTANTE IMPUTADO:")
print(datos)
print("\n")

# ============================================================
# 4. TRANSFORMAR CIUDAD CON LABEL ENCODING
# ============================================================

print("=" * 60)
print("LABEL ENCODING")
print("=" * 60)

label_encoder = LabelEncoder()
datos['Ciudad_Label'] = label_encoder.fit_transform(datos['Ciudad'])

print("Correspondencia (Label Encoding):")
for i, ciudad in enumerate(label_encoder.classes_):
    print(f"  {ciudad} → {i}")

print("\nDATOS CON LABEL ENCODING:")
print(datos[['ClienteID', 'Ciudad', 'Ciudad_Label']])
print("\n")

# ============================================================
# 5. TRANSFORMAR CIUDAD CON ONE-HOT ENCODING (PANDAS)
# ============================================================

print("=" * 60)
print("ONE-HOT ENCODING (con Pandas)")
print("=" * 60)

onehot_df = pd.get_dummies(datos['Ciudad'], prefix='Ciudad')
datos_onehot = pd.concat([datos, onehot_df], axis=1)

print("DATOS CON ONE-HOT ENCODING:")
print(datos_onehot[['ClienteID', 'Ciudad', 'Ciudad_Barcelona', 'Ciudad_Madrid', 'Ciudad_Sevilla']])
print("\n")

# ============================================================
# 6. TRANSFORMAR CIUDAD CON ONE-HOT ENCODING (SCIKIT-LEARN)
# ============================================================

print("=" * 60)
print("ONE-HOT ENCODING (con Scikit-Learn)")
print("=" * 60)

onehot_encoder = OneHotEncoder(sparse_output=False)
ciudades_ohe = onehot_encoder.fit_transform(datos[['Ciudad']])
ciudades_df = pd.DataFrame(ciudades_ohe, columns=onehot_encoder.get_feature_names_out(['Ciudad']))
datos_ohe_sklearn = pd.concat([datos, ciudades_df], axis=1)

print("DATOS CON ONE-HOT ENCODING (Scikit-Learn):")
print(datos_ohe_sklearn[['ClienteID', 'Ciudad', 'Ciudad_Barcelona', 'Ciudad_Madrid', 'Ciudad_Sevilla']])
print("\n")

# ============================================================
# 7. APLICAR VARIABLES DUMMY PARA CIUDAD
# ============================================================

print("=" * 60)
print("VARIABLES DUMMY (con drop_first=True)")
print("=" * 60)

dummies_df = pd.get_dummies(datos['Ciudad'], prefix='Ciudad', drop_first=True)
datos_dummies = pd.concat([datos, dummies_df], axis=1)

print("DATOS CON VARIABLES DUMMY (Barcelona es la categoria de referencia):")
print(datos_dummies[['ClienteID', 'Ciudad', 'Ciudad_Madrid', 'Ciudad_Sevilla']])
print("\nInterpretacion: Si ambas columnas son 0, el cliente es de Barcelona")
print("\n")

# ============================================================
# 8. ESCALAR COLUMNAS EDAD E INGRESOS
# ============================================================

columnas_a_escalar = ['Edad', 'Ingresos']

# 8.1 NORMALIZACION MIN-MAX
print("=" * 60)
print("NORMALIZACION MIN-MAX (valores entre 0 y 1)")
print("=" * 60)

minmax_scaler = MinMaxScaler()
datos_minmax = datos.copy()
datos_minmax[columnas_a_escalar] = minmax_scaler.fit_transform(datos[columnas_a_escalar])

print(datos_minmax)
print("\n")

# 8.2 ESTANDARIZACION Z-SCORE
print("=" * 60)
print("ESTANDARIZACION Z-SCORE (media=0, desviacion=1)")
print("=" * 60)

standard_scaler = StandardScaler()
datos_standard = datos.copy()
datos_standard[columnas_a_escalar] = standard_scaler.fit_transform(datos[columnas_a_escalar])

print(datos_standard)
print("\n")

# ============================================================
# 9. COMPARACION DE ESCALAMIENTO POR CLIENTE
# ============================================================

print("=" * 60)
print("COMPARACION DE ESCALAMIENTO PARA CADA CLIENTE")
print("=" * 60)

print("\nCliente 1 (Edad=25, Ingresos=30000):")
print(f"  Original: Edad=25, Ingresos=30000")
print(f"  Min-Max: Edad={datos_minmax.loc[0, 'Edad']:.3f}, Ingresos={datos_minmax.loc[0, 'Ingresos']:.3f}")
print(f"  Z-Score: Edad={datos_standard.loc[0, 'Edad']:.3f}, Ingresos={datos_standard.loc[0, 'Ingresos']:.3f}")

print("\nCliente 2 (Edad=45, Ingresos=50000):")
print(f"  Original: Edad=45, Ingresos=50000")
print(f"  Min-Max: Edad={datos_minmax.loc[1, 'Edad']:.3f}, Ingresos={datos_minmax.loc[1, 'Ingresos']:.3f}")
print(f"  Z-Score: Edad={datos_standard.loc[1, 'Edad']:.3f}, Ingresos={datos_standard.loc[1, 'Ingresos']:.3f}")

print("\nCliente 3 (Edad=30, Ingresos=40000 - imputado):")
print(f"  Original: Edad=30, Ingresos=40000")
print(f"  Min-Max: Edad={datos_minmax.loc[2, 'Edad']:.3f}, Ingresos={datos_minmax.loc[2, 'Ingresos']:.3f}")
print(f"  Z-Score: Edad={datos_standard.loc[2, 'Edad']:.3f}, Ingresos={datos_standard.loc[2, 'Ingresos']:.3f}")

print("\nCliente 4 (Edad=40, Ingresos=40000):")
print(f"  Original: Edad=40, Ingresos=40000")
print(f"  Min-Max: Edad={datos_minmax.loc[3, 'Edad']:.3f}, Ingresos={datos_minmax.loc[3, 'Ingresos']:.3f}")
print(f"  Z-Score: Edad={datos_standard.loc[3, 'Edad']:.3f}, Ingresos={datos_standard.loc[3, 'Ingresos']:.3f}")
print("\n")

# ============================================================
# 10. CREAR DATASET FINAL CON TODAS LAS TRANSFORMACIONES
# ============================================================

print("=" * 60)
print("DATASET TRANSFORMADO FINAL")
print("=" * 60)

dataset_final = datos.copy()
dataset_final['Ingresos_imputados'] = datos['Ingresos']
dataset_final['Ciudad_Label'] = label_encoder.fit_transform(datos['Ciudad'])
dataset_final = pd.concat([dataset_final, pd.get_dummies(datos['Ciudad'], prefix='Ciudad')], axis=1)
dataset_final[['Edad_MinMax', 'Ingresos_MinMax']] = minmax_scaler.fit_transform(datos[columnas_a_escalar])
dataset_final[['Edad_ZScore', 'Ingresos_ZScore']] = standard_scaler.fit_transform(datos[columnas_a_escalar])

print(dataset_final)
print("\n")

# ============================================================
# 11. GUARDAR DATASET TRANSFORMADO A CSV
# ============================================================

dataset_final.to_csv('datos_preprocesados.csv', index=False)
print("=" * 60)
print("DATASET TRANSFORMADO GUARDADO: 'datos_preprocesados.csv'")
print("=" * 60)

# ============================================================
# 12. MOSTRAR RESUMEN DE TRANSFORMACIONES
# ============================================================

print("\n")
print("=" * 60)
print("RESUMEN DE TRANSFORMACIONES APLICADAS")
print("=" * 60)
print("""
1. IMPUTACION: Se relleno el valor nulo de Ingresos con la media (40000)

2. LABEL ENCODING: Barcelona→0, Madrid→1, Sevilla→2

3. ONE-HOT ENCODING: Se crearon columnas binarias por cada ciudad

4. VARIABLES DUMMY: Se elimino una columna para evitar multicolinealidad

5. NORMALIZACION MIN-MAX: Edad e Ingresos escalados al rango [0,1]

6. ESTANDARIZACION Z-SCORE: Edad e Ingresos con media=0 y desviacion=1
""")


# ============================================================
# CODIGO PARA CREAR EL DATASET TRANSFORMADO (alternativo)
# ============================================================

# Crear el dataset final con todas las columnas
scaled_minmax_alt = minmax_scaler.transform(datos[columnas_a_escalar])
scaled_standard_alt = standard_scaler.transform(datos[columnas_a_escalar])

dataset_final = pd.DataFrame({
    'ClienteID': datos['ClienteID'],
    'Edad': datos['Edad'],
    'Ciudad': datos['Ciudad'],
    'Ingresos': datos['Ingresos'],
    'Ingresos_imputados': datos['Ingresos'],
    'Ciudad_Label': label_encoder.transform(datos['Ciudad']),
    'Ciudad_Barcelona': [1 if c == 'Barcelona' else 0 for c in datos['Ciudad']],
    'Ciudad_Madrid': [1 if c == 'Madrid' else 0 for c in datos['Ciudad']],
    'Ciudad_Sevilla': [1 if c == 'Sevilla' else 0 for c in datos['Ciudad']],
    'Edad_MinMax': scaled_minmax_alt[:, 0],
    'Ingresos_MinMax': scaled_minmax_alt[:, 1],
    'Edad_ZScore': scaled_standard_alt[:, 0],
    'Ingresos_ZScore': scaled_standard_alt[:, 1]
})

# Mostrar
print(dataset_final)

# Guardar a CSV
dataset_final.to_csv('datos_preprocesados.csv', index=False)
print("Archivo guardado: datos_preprocesados.csv")
```
