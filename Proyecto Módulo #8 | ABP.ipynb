```python
# ====================================================
# Proyecto Módulo 8: Clasificador de Ropa con Deep Learning
# Dataset: Fashion-MNIST (10 tipos de prendas)
# Frameworks: TensorFlow / Keras
# ====================================================

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.utils import to_categorical

# Semilla para reproducibilidad
tf.random.set_seed(42)
np.random.seed(42)

# ====================================================
# 1. Carga y preprocesamiento de datos
# ====================================================
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()

# Nombres de las clases
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']

# Normalizar píxeles (0-255 -> 0-1)
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0

# Para la red densa: aplanar imágenes 28x28 -> 784
x_train_flat = x_train.reshape(-1, 28*28)
x_test_flat = x_test.reshape(-1, 28*28)

# Para la CNN: añadir canal de color (escala de grises -> 1)
x_train_cnn = x_train.reshape(-1, 28, 28, 1)
x_test_cnn = x_test.reshape(-1, 28, 28, 1)

# One-hot encoding de etiquetas
y_train_cat = to_categorical(y_train, 10)
y_test_cat = to_categorical(y_test, 10)

print("Datos cargados:")
print(f"  Entrenamiento: {x_train.shape}, Test: {x_test.shape}")
print(f"  Entrenamiento flat: {x_train_flat.shape}")
print(f"  Entrenamiento CNN: {x_train_cnn.shape}")

# Visualización rápida de algunas imágenes
plt.figure(figsize=(10,5))
for i in range(10):
    plt.subplot(2,5,i+1)
    plt.imshow(x_train[i], cmap='gray')
    plt.title(class_names[y_train[i]])
    plt.axis('off')
plt.tight_layout()
plt.show()

# ====================================================
# 2. Red Neuronal Densa (MLP)
# ====================================================
def build_dense_model():
    model = models.Sequential([
        layers.Dense(256, activation='relu', input_shape=(784,)),
        layers.Dropout(0.3),
        layers.Dense(128, activation='relu'),
        layers.Dropout(0.3),
        layers.Dense(64, activation='relu'),
        layers.Dense(10, activation='softmax')
    ])
    return model

dense_model = build_dense_model()
dense_model.compile(optimizer='adam',
                    loss='categorical_crossentropy',
                    metrics=['accuracy'])

print("\nEntrenando red densa...")
history_dense = dense_model.fit(x_train_flat, y_train_cat,
                                epochs=20,
                                batch_size=128,
                                validation_split=0.2,
                                verbose=1)

# Evaluación red densa
test_loss_dense, test_acc_dense = dense_model.evaluate(x_test_flat, y_test_cat, verbose=0)
print(f"\nPrecisión en test (red densa): {test_acc_dense:.4f}")

# Curvas de aprendizaje red densa
plt.figure()
plt.plot(history_dense.history['accuracy'], label='Entrenamiento')
plt.plot(history_dense.history['val_accuracy'], label='Validación')
plt.title('Precisión - Red densa')
plt.xlabel('Épocas')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

# ====================================================
# 3. Red Neuronal Convolutiva (CNN)
# ====================================================
def build_cnn_model():
    model = models.Sequential([
        layers.Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)),
        layers.MaxPooling2D((2,2)),
        layers.Conv2D(64, (3,3), activation='relu'),
        layers.MaxPooling2D((2,2)),
        layers.Conv2D(128, (3,3), activation='relu'),
        layers.Flatten(),
        layers.Dropout(0.5),
        layers.Dense(128, activation='relu'),
        layers.Dense(10, activation='softmax')
    ])
    return model

cnn_model = build_cnn_model()
cnn_model.compile(optimizer='adam',
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])

# Early stopping
early_stop = keras.callbacks.EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

print("\nEntrenando CNN...")
history_cnn = cnn_model.fit(x_train_cnn, y_train_cat,
                            epochs=30,
                            batch_size=64,
                            validation_split=0.2,
                            callbacks=[early_stop],
                            verbose=1)

# Evaluación CNN
test_loss_cnn, test_acc_cnn = cnn_model.evaluate(x_test_cnn, y_test_cat, verbose=0)
print(f"\nPrecisión en test (CNN): {test_acc_cnn:.4f}")
print(f"Mejora respecto a red densa: {(test_acc_cnn - test_acc_dense)*100:.2f}%")

# Curvas de aprendizaje CNN
plt.figure()
plt.plot(history_cnn.history['accuracy'], label='Entrenamiento CNN')
plt.plot(history_cnn.history['val_accuracy'], label='Validación CNN')
plt.title('Precisión - CNN')
plt.xlabel('Épocas')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

# ====================================================
# 4. Matriz de confusión y métricas detalladas
# ====================================================
y_pred_probs = cnn_model.predict(x_test_cnn)
y_pred = np.argmax(y_pred_probs, axis=1)

cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(10,8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=class_names, yticklabels=class_names)
plt.title('Matriz de confusión - CNN')
plt.xlabel('Predicción')
plt.ylabel('Real')
plt.show()

print("\nReporte de clasificación - CNN:")
print(classification_report(y_test, y_pred, target_names=class_names))

# ====================================================
# 5. Predicción sobre imágenes nuevas (simulación)
# ====================================================
# Tomamos una imagen del test set que no se usó en entrenamiento
sample_index = 0  # puedes cambiar el índice
img_test = x_test_cnn[sample_index]
true_label = y_test[sample_index]

pred_prob = cnn_model.predict(img_test.reshape(1,28,28,1))
pred_class = np.argmax(pred_prob)

plt.figure()
plt.imshow(x_test[sample_index], cmap='gray')
plt.title(f"Real: {class_names[true_label]} | Predicción: {class_names[pred_class]} (prob: {pred_prob[0][pred_class]:.2f})")
plt.axis('off')
plt.show()

# Función para predecir desde archivo externo (descomentar si se tiene imagen local)
"""
from PIL import Image
def predecir_imagen(ruta_imagen):
    img = Image.open(ruta_imagen).convert('L')  # escala de grises
    img = img.resize((28,28))
    img_array = np.array(img) / 255.0
    img_array = img_array.reshape(1,28,28,1)
    pred = cnn_model.predict(img_array)
    clase = np.argmax(pred)
    print(f"Clase predicha: {class_names[clase]} (prob: {pred[0][clase]:.2f})")
    plt.imshow(img, cmap='gray')
    plt.title(class_names[clase])
    plt.axis('off')
    plt.show()
"""

print("\nProyecto completado exitosamente.")

# ====================================================
# Visualizaciones del Proyecto: Clasificador de Ropa
# ====================================================

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report
from tensorflow.keras.datasets import fashion_mnist

# Cargar datos (solo para visualizar, sin necesidad de entrenar)
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']

# Normalizar para visualización correcta (opcional)
x_train = x_train / 255.0

# ====================================================
# 1. Visualización de muestras del dataset
# ====================================================
plt.figure(figsize=(10,5))
for i in range(10):
    plt.subplot(2,5,i+1)
    plt.imshow(x_train[i], cmap='gray')
    plt.title(class_names[y_train[i]])
    plt.axis('off')
plt.suptitle('Ejemplos de imágenes del dataset Fashion-MNIST', fontsize=14)
plt.tight_layout()
plt.show()

# ====================================================
# 2. Distribución de clases (opcional)
# ====================================================
unique, counts = np.unique(y_train, return_counts=True)
plt.figure(figsize=(10,5))
plt.bar(class_names, counts, color='skyblue')
plt.title('Distribución de clases en entrenamiento')
plt.xlabel('Categoría')
plt.ylabel('Número de imágenes')
plt.xticks(rotation=45)
plt.show()

# ====================================================
# 3. Curvas de aprendizaje (precisión y pérdida)
# Asumiendo que ya tienes history_dense y history_cnn de los entrenamientos
# ====================================================
# Ejemplo con history_dense (red densa) y history_cnn (CNN)

# history_dense = dense_model.fit(...)   # obtenido del entrenamiento
# history_cnn = cnn_model.fit(...)

# Curvas de precisión
plt.figure(figsize=(12,4))

plt.subplot(1,2,1)
plt.plot(history_dense.history['accuracy'], label='Entrenamiento Densa')
plt.plot(history_dense.history['val_accuracy'], label='Validación Densa')
plt.plot(history_cnn.history['accuracy'], label='Entrenamiento CNN', linestyle='--')
plt.plot(history_cnn.history['val_accuracy'], label='Validación CNN', linestyle='--')
plt.title('Precisión durante el entrenamiento')
plt.xlabel('Épocas')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(1,2,2)
plt.plot(history_dense.history['loss'], label='Entrenamiento Densa')
plt.plot(history_dense.history['val_loss'], label='Validación Densa')
plt.plot(history_cnn.history['loss'], label='Entrenamiento CNN', linestyle='--')
plt.plot(history_cnn.history['val_loss'], label='Validación CNN', linestyle='--')
plt.title('Pérdida durante el entrenamiento')
plt.xlabel('Épocas')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()

# ====================================================
# 4. Matriz de confusión de la CNN
# ====================================================
# y_test son etiquetas reales, y_pred son predicciones del modelo CNN
# y_pred = np.argmax(cnn_model.predict(x_test_cnn), axis=1)

cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(10,8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=class_names, yticklabels=class_names)
plt.title('Matriz de confusión - CNN sobre Fashion-MNIST')
plt.xlabel('Predicción')
plt.ylabel('Valor real')
plt.show()

# ====================================================
# 5. Ejemplo de predicción con imagen individual
# ====================================================
# Tomamos una imagen del test set
index = 0   # puedes cambiar el índice
imagen = x_test[index]
real = y_test[index]
pred_prob = cnn_model.predict(imagen.reshape(1,28,28,1))
pred = np.argmax(pred_prob)

plt.figure(figsize=(4,4))
plt.imshow(imagen, cmap='gray')
plt.title(f'Real: {class_names[real]}\nPredicción: {class_names[pred]} (prob: {pred_prob[0][pred]:.2f})')
plt.axis('off')
plt.show()

# ====================================================
# 6. Comparación de precisión final (gráfico de barras)
# ====================================================
acc_dense = test_acc_dense   # ejemplo: 0.88
acc_cnn = test_acc_cnn       # ejemplo: 0.925

plt.figure(figsize=(6,5))
plt.bar(['Red densa (MLP)', 'CNN'], [acc_dense, acc_cnn], color=['orange', 'green'])
plt.ylim([0,1])
plt.ylabel('Precisión en test')
plt.title('Comparación de precisión entre modelos')
for i, v in enumerate([acc_dense, acc_cnn]):
    plt.text(i, v + 0.01, f'{v:.3f}', ha='center')
plt.show()
```
